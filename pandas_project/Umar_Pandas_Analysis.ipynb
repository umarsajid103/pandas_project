Read a CSV file, inspect the data, compute summary statistics, filter and slice data, and save the processed output using Pandas.

```python
## Import Libraries
```

```python
import pandas as pd
```

```python
## Read Dataset
```

```python
df = pd.read_csv("data.csv")
df
```

```python
## Inspect the Data
```

```python
print("First 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

print("\nData types:")
print(df.dtypes)

print("\nDataset info:")
df.info()
```

```python
## Summary Statistics
```

```python
print("Mean:")
print(df.mean(numeric_only=True))

print("\nMedian:")
print(df.median(numeric_only=True))

print("\nMinimum:")
print(df.min(numeric_only=True))

print("\nMaximum:")
print(df.max(numeric_only=True))

print("\nCount:")
print(df.count())
```

```python
## Filter Rows (Age > 25)
```

```python
filtered_rows = df[df["Age"] > 25]
filtered_rows
```

```python
## Select Specific Columns
```

```python
selected_columns = df[["Name", "Salary"]]
selected_columns
```

```python
## Slice Subset (First 3 Rows)
```

```python
subset = df.iloc[0:3]
subset
```

```python
## Save Filtered Output
```

```python
filtered_rows.to_csv("filtered_output.csv", index=False)
print("Filtered file saved successfully!")
```

## Conclusion

In this project, we successfully:

- Loaded CSV data into a Pandas DataFrame  
- Inspected the dataset structure  
- Generated summary statistics  
- Filtered and sliced the data  
- Exported the processed results  

This demonstrates basic data analysis workflow using Pandas.

## Data Visualization

The following visualization shows the relationship between employee names and their salaries.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Salary"])
plt.title("Salary Distribution by Employee")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
